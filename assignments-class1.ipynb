```python
# Book exercise 7.For the function f (x) = 3x + 1 what is the slope at x = 3?

from sympy import *
x = symbols('x')

#f(x)=3x**2+1
f = 3 * x ** 2 + 1
plot(f)

#Slope at x = 3 
dx_f = diff(f)
print(dx_f.subs(x,3))
```
