```python
# create own function with 'def' in example the functionis named f(x)
# to show the result x in x_values print(y) = y = f(x)
# solves return for all x_values in list

def f(x):
    return 2 * x + 1

x_values = [0, 1, 2, 3]

for x in x_values:
    y = f(x)
    print(y)
```

```python
# packages - import packages, package can be imported by 'import' just as an app,
# the math package does include square root i.e. 

import math #does import the whole package
math.sqrt(7921)

from math import sqrt #does only use the square root from the import package (less CPU used)
sqrt(7921)
```

```python
## plotting functions by sympy

from sympy import symbols
from sympy.plotting import plot

x = symbols ('x') 
f = x**3 + 15*x + 300
plot(f)
```

```python
# presentation example
from sympy import symbols 
from sympy.plotting import plot

x = symbols ('x') 
f = 2*x+1
plot(f)
```
